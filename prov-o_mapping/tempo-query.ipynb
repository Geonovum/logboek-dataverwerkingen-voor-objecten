### notebook om opentelemetry trace data uit een grafana tempo database op te halen



```python
import requests 
import pandas as pd
import json
```

```python

url = "http://localhost:3200/api/search?q={}&limit=20"
response = requests.get(url)

if response.status_code == 200:
    result = response.json()
```

```python
traces = result['traces'][0]

traceid = traces['traceID']
serviceName = traces['rootServiceName']

print(traceid, serviceName)
```

```python
url = "http://localhost:3200/api/traces/" + str(traceid)
response = requests.get(url)

if response.status_code == 200:
    traceresult = response.json()
else:
    print(f"Error: {response.status_code, response.text}")

traceresult
```
